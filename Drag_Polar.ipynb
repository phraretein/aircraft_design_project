# Drag Polar

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cl = np.arange(start=0.0, stop=2.05, step=0.05)
```

```python
cl
```

```python
cd_0 = 0.00825
cd_0_to = 0.023473033
cd_0_l = 0.108879561
```

```python
K = 0.041422059
K_to = 0.044152716
K_l = 0.047267005
```

```python
cd = cd_0 + K*(cl**2)
```

```python
cd
```

```python
cd_to = cd_0_to + K_to*(cl**2)
cd_to
```

```python
cd_l = cd_0_l + K_l*(cl**2)
cd_l
```

```python
plt.plot(cd, cl, 'midnightblue', label='Cruise condition')
plt.plot(cd_to, cl, 'forestgreen', label='Takeoff condition')
plt.plot(cd_l, cl, 'firebrick', label='Landing condition')
plt.xlim(0, 0.2)
plt.ylim(0, 2.1)
plt.legend(bbox_to_anchor=(1, 0.7), loc='upper left', prop={'size':13})
params = {'mathtext.default': 'regular'}  
plt.rcParams.update(params)
plt.xlabel("$C_{d}\ [-]$", fontsize=12)
plt.ylabel("$C_{l}\ [-]$", fontsize=12)
plt.title("Drag Polar Estimation", fontsize=20)
plt.grid(b=None, which='major', axis='both')
```
